# YOLO Training + Prediction Workflow

## 1. Import the required libraries
This step establishes the computational framework by importing the YOLO module for object detection and the operating system interface for file management. It enables access to high-level abstractions for model definition, training, and inference, while also allowing interaction with the local file system where datasets, model weights, and outputs are stored. Conceptually, this initializes the software environment necessary for executing a deep learning pipeline.

```python
from ultralytics import YOLO
import os
```

## 2. Load the base YOLO model
Here, a pretrained YOLO model (e.g., yolov8n.pt) is instantiated, which initializes the network with weights learned from large-scale datasets. These pretrained weights encode generalized visual features such as edges, textures, and object structures. This process leverages transfer learning, significantly reducing training time and improving convergence compared to training from random initialization.

```python
model = YOLO("yolov8n.pt")
```

## 3. Train the model
During training, the model is exposed to a labeled dataset (e.g., coco8.yaml) over multiple epochs. In each iteration, the model performs forward propagation to generate predictions and backward propagation to adjust its weights based on the computed loss. Hyperparameters such as image size standardize input dimensions, while repeated exposure to data refines the model’s ability to localize objects and classify them accurately. The training process produces multiple checkpoints, including the best-performing weights.

```python
model.train(
    data="coco8.yaml",   # tiny dataset (8 images)
    epochs=10,
    imgsz=416
)
```

## 4. Check the current working directory

```python
print(os.getcwd())
```

 ## 5. Load your trained model
 After training, the model is reloaded using the saved weight file (typically best.pt), which contains the optimized parameters corresponding to the highest validation performance. This step transitions the model from a training state to an inference-ready state, ensuring that subsequent predictions utilize the most effective learned representation.

```python
model = YOLO("/Users/shhivani/Desktop/my-yolo-project/runs/detect/train/weights/best.pt")
```

## 6. Run prediction on an image
In this phase, the trained model is applied to static image data. The model performs a forward pass, producing bounding boxes, class labels, and confidence scores for detected objects. A confidence threshold filters low-probability detections, and the results may be visualized and saved. This step demonstrates the model’s ability to generalize learned features to unseen data.

```python
model.predict(
    source="https://ultralytics.com/images/bus.jpg",
    save=True,
    conf=0.25
)
```

## 7. Run prediction using your webcam
Finally, the inference process is extended to a real-time input stream from a webcam. The model continuously processes frames, applying detection in near real-time. This step illustrates deployment in a dynamic environment, where the model must maintain performance under varying conditions such as motion, lighting, and scene complexity, thereby completing the transition from training to practical application.

```python
model.predict(source=0, show=True)
```

The workflow begins with the import of the necessary libraries, which establishes access to the YOLO framework and system-level file management, forming the computational environment. Subsequently, a pretrained YOLO model is loaded, initializing the network with learned weights derived from large-scale datasets, thereby enabling efficient transfer learning rather than training from random initialization. The model is then trained on a specified dataset, during which its weights are iteratively updated across multiple epochs through backpropagation to minimize detection loss, while image resizing ensures consistent input dimensionality. The working directory is verified to establish the file system context in which all training artifacts—such as logs, metrics, and weight checkpoints—are stored. Following training, the optimal set of learned parameters, typically saved as best.pt, is reloaded to ensure that subsequent inference uses the highest-performing model state. The trained model is then applied to image data, where it performs forward propagation to generate bounding boxes, class predictions, and associated confidence scores, which may be optionally persisted for analysis. Finally, the same inference mechanism is extended to a live webcam stream, enabling real-time object detection, thereby demonstrating the transition from static evaluation to continuous deployment in a dynamic input setting.
